# 결측치 및 이상치 처리

```python
df = pd.read_csv('carsharing.csv', encoding = 'euc-kr')
```

```python
df.info()
```

```python
#칼럼명 정리
df = df.rename({'예약시간 (시)' : '예약시간_시', '예약대기시간 (시)' : '예약대기시간_시', '이용시간 (시)' : '이용시간_시', '최초 탑승일시' : '최초탑승일시','최근 탑승일시' : '최근탑승일시'}, axis = 'columns')
```

```python
df.info()
```

거주단지가 관리자이거나 탈퇴회원인 데이터 제거

```python
df = df[~((df['거주단지'] == '★관리자')|(df['거주단지'] == '☆탈퇴회원'))]
df.shape
```

등급이 null값인 데이터 제거

```python
df = df.dropna(axis = 'index', how = 'any', subset =['등급'])
```

성별이 이상한 데이터 2개 처리하기

```python
df['성별'].value_counts()
```

```python
df[~((df['성별']=='남')|(df['성별']=='여'))]
```

```python
#성별 확인
#df.query('아이디 == "gthasu"')
#df.query('아이디 == "skykoko1385"')
```

```python
df['성별'] = df['성별'].apply(lambda x : '남' if x == '남' else ('여' if x == '여' else '남'))
```

```python
df['성별'].value_counts()
```

예약대기시간_시가 -인 값 제거

총요금이 - 인 값 제거 (환불인지는 잘 모르겠음)

```python
df = df[df['총요금'] >= 0]
df.shape
```

예약요금 "59080" 인 고객의 경우 쿠폰 사용으로 0으로 대치\
반납지연요금 "75680"인 고객의 경우 75680으로 대치

```python
df.query('아이디 == "yunki68"')
```

```python
def cleansing(x):
    clean_x = re.sub(r"[^a-zA-Z0-9]","",str(x))
    return clean_x
```

```python
#""제거
df['예약요금'] = df['예약요금'].apply(cleansing)
```

```python
# 쿠폰사용, 계산하면 예약 요금 0
df['예약요금'].iloc[3931:3932] = 0
```

반납지연요금 "75680"인 고객의 경우 75680으로 대치

```python
df['반납지연요금'] = df['반납지연요금'].apply(cleansing)
```

```python
#예약요금, 반납지연요금 object -> 정수, 실수형으로
df = df.astype({'예약요금' : 'int','반납지연요금' : 'float'})
df.info()
```

```python
#인덱스 초기화
df = df.reset_index().iloc[:,1:]
```

```python
df.to_csv('carsharing_checkpoint.csv', index=False,  encoding = 'euc-kr')
```
